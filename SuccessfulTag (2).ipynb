```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
df= pd.read_excel("RFMInew.xlsx")
```

```python
df
```

```python
df.rename(columns={'What questions do you want answered?':'Query'}, inplace=True)
```

```python
df
```

```python
faqtag = {'Scholarship': ['Scholarship', 'financial aid', 'financial assistance','monetary aid','fellowship','grant'],
          'Test Score': ['toefl', 'ielts','scores','score','gre','eligiblity'],
          'Housing': ['accommodation','housing','off-campus','locations','lease','rent','budget','living'],
          'Co-op': ['coop','co-op','internships','internship','placements','job','eligibility','experience'],
          'Admission':['application','deadline','admission','ogs','uhcs','health report','portal','visa','I-20','i20']
         }
```

```python
def tagApply(x):
    tagList = []
    for key, values in faqtag.items():
        for value in values:
            if value in x.lower():
                if key not in tagList:
                    tagList.append(key)
    if len(tagList) == 0:
        tagList.append("Misc")
    tagString = ", ".join(tagList)
    return tagString
```

```python
df['Tag']=df['Query'].apply(tagApply)
```

```python
df
```

```python
tagCount = Counter(", ".join(df['Tag']).split(", "))
tagCount
```

```python
tagCountSorted = sorted(tagCount, key=tagCount.get, reverse=True)
tagCountSorted
```

```python
tagdict=dict(tagCount)
```

```python
df2=pd.DataFrame.from_dict(tagdict,orient="index").reset_index()
```

```python
df2 = df2.rename(columns={'index':'Query Type', 0:'Count'})
df2
```
